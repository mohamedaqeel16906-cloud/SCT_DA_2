```python
import pandas as pd

df = pd.read_csv("../data/Global_Superstore.csv", encoding="latin1")
```

```python
df.head()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df["Postal Code"] = df["Postal Code"].fillna(df["Postal Code"].median())
```

```python
df.isnull().sum()
```

```python
df.duplicated().sum()
```

```python
df = df.drop_duplicates()
```

```python
df["Order Date"] = pd.to_datetime(df["Order Date"])
df["Ship Date"] = pd.to_datetime(df["Ship Date"])
```

```python
df.dtypes
```

```python
df.info()
```

```python
df.to_csv("../cleaned_data/Global_Superstore_Cleaned.csv", index=False)
```
